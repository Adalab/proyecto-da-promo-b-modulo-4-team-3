Hemos utilizado los archivos disponibles en la base de datos de Kaggle de https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset
En particular el archivo:
"credits.csv": Consists of Cast and Crew Information for all our movies. Available in the form of a stringified JSON Object.
Del que hemos obtenido información sobre la distribución de las mujeres en las principales categorías de la industria del cine.
Este archivo, al convertilo a un dataframe, tiene 3 columnas:
cast: información sobre los actrores/actrices de la película (que no hemos utilizado)
crew: información sobre el personal que participa en realización de la película: Dirección, producción, guión, etc.
id: el id de la película

```python
import ast

# Paso 1: Convertir las cadenas de texto de la columna 'cast' en listas de diccionarios
df['cast'] = df['cast'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df["cast"][0]
```

```python
# Paso 2: Expandimos la columna 'cast' para que cada diccionario en la lista tenga su propia fila
df_cast_exploded = df.explode('cast').reset_index(drop=True)

df_cast_exploded.head(1)

```

```python
# Paso 3: Normalizamos la columna 'cast' (que contiene diccionarios) en columnas separadas
df_cast_final = pd.json_normalize(df_cast_exploded['cast'])

df_cast_final.head(1)
```

```python
df_cast_final.shape
```

```python
# Paso 4: Añadir la columna 'id' para mantener la relación con el DataFrame original
df_cast_final['id'] = df_cast_exploded['id']

# Mostramos el resultado
df_cast_final.head(1)
```
